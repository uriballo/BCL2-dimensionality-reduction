# Matrices

```julia
adjacency_file = matopen("./matrices/adjacency.mat")
cluster1_file = matopen("./matrices/S1.mat")
cluster2_file = matopen("./matrices/S2.mat")
cluster3_file = matopen("./matrices/S3.mat")
```

```julia
S = read(adjacency_file, "S")
S1 = read(cluster1_file, "S1")
S2 = read(cluster2_file, "S2")
S3 = read(cluster3_file, "S3")

s_dim = size(S)
s1_dim = size(S1)
s2_dim = size(S2)
s3_dim = size(S3)

println("DIMENSIONS")
println("\tS: ", s_dim)
println("\tS1: ", s1_dim)
println("\tS2: ", s2_dim)
println("\tS3: ", s3_dim)
```
