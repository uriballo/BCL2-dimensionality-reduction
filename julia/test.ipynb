```python
import sympy as sp
sp.init_printing()

# Define the variables
x1, x2, x3, A1, A2, A3,  B = sp.symbols('x_1 x_2 x_3 A_1 A_2 A_3 B')

# Define the constants
mu11 = 16273.465487425427
mu22 = 7800.653148254713
mu33 = 140.00867213154856

mu12 = 8.841868906761077e-7
mu13 = 0.0
mu23 = 11.674613405026601

mu21 = 0.0583254283653529
mu31 = 0.0
mu32 = 23.34605808990502

# Define the system of equations
eq1 = A1 - B*x1 - mu11*x1**2 - x1*mu12*x2
eq2 = 0 - B*x2 - x2*mu21*x1 - mu22*x2**2 - x2*mu23*x3
eq3 = 0 - B*x3 - x3*mu32*x2 - mu33*x3**2

# Solve the system of equations
solution = sp.solve([eq1, eq2, eq3], (x1, x2, x3), dict=True)
```

```python
if not solution:
    print("No solutions found.")
else:
    for sol in solution:
        print(sol)
```
